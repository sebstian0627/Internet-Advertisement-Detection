```python
def scorer(s1,s2):
  embedding1=model.encode(s1, convert_to_tensor=True)
  embedding2 = model.encode(s2, convert_to_tensor=True)
  cosine_scores = util.pytorch_cos_sim(embedding1, embedding2)
  return cosine_scores.item()
```
